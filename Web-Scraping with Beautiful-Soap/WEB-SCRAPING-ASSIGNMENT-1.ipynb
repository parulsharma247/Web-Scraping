<h3>i) Top 10 ODI teams in men’s cricket along with the records for matches, points and rating.

```python
page = requests.get("https://www.icc-cricket.com/rankings/mens/team-rankings/odi")
soup = BeautifulSoup(page.content)
soup
```

#### getting the teams name

```python
teams_name = soup.find_all('span', class_='u-hide-phablet')
print(teams_name[0:5])

teams_list =[]

for team in teams_name[0:10]:
    teams_list.append(team.text)
    
teams_list
```

#### getting the no of matches and points

```python
matches_nz = soup.find_all('td', class_='rankings-block__banner--matches')
points_nz = soup.find_all('td', class_='rankings-block__banner--points')

matches = soup.find_all('td', class_='table-body__cell u-center-text')
print(matches_nz,points_nz, '\n', )
matches[0:8]
```

```python
matches_list =[]
points_list = []

for no in matches_nz:
    matches_list.append(no.text)
print('matches_nz_played = ', matches_list)

for points in points_nz:
    points_list.append(points.text)
print('points of NZ = ',points_list)



for count,no in enumerate(matches[0:18]):
    #print(count, no)
    if count%2==0:
        matches_list.append(no.text)
        #print(no.text)
    else: # points of each team
        points_list.append(no.text)
        
print('\n',points_list)
matches_list
```

<h4>Getting the teams rating

```python
nz_rating = soup.find_all('td', class_='rankings-block__banner--rating u-text-right')
other_teams_rating = soup.find_all('td',class_='table-body__cell u-text-right rating')
nz_rating, other_teams_rating[0:9]
```

```python
rating_list = []

for rating in nz_rating:
    rating_list.append(rating.text.replace('\n','').replace(' ',''))
    
for rating in other_teams_rating[0:9]:
    rating_list.append(rating.text)
    
rating_list
```

```python
len(teams_list), len(matches_list),len(points_list), len(rating_list)
```

```python
top_10_icc_odi_teams = pd.DataFrame()

top_10_icc_odi_teams['Teams'] =  teams_list
top_10_icc_odi_teams['Matches']  = matches_list
top_10_icc_odi_teams['Points']  = points_list
top_10_icc_odi_teams['Rating'] = rating_list 

top_10_icc_odi_teams
```

<h3>Task 7 : Write a python program to scrape details of all the mobile phones under Rs. 20,000 
listed on Amazon.in. The scraped data should include Product Name, Price, Image URL 
and Average Rating.


```python
# https://www.amazon.in/s?k=mobile+phones+under+20000&ref=nav_signin

page = requests.get("https://www.amazon.in/s?k=mobile+phones+under+20000&s=relevanceblender&qid=1620576461&ref=sr_st_relevanceblender")
soup = BeautifulSoup(page.content)
soup
```

```python
# getting the product names

products_list = []
product_names = soup.find_all('span', class_='a-size-medium a-color-base a-text-normal')
#print(product_names)

for name in product_names:
    products_list.append(name.text)

print(len(products_list))    
products_list
```

```python
# getting the product prices

prices_list = []
product_prices = soup.find_all('span', class_='a-price-whole')
# product_prices
for price in product_prices:
    prices_list.append(price.text)

print(len(prices_list[4:]))
prices_list=prices_list[4:]
prices_list
```

```python
# getting the Images URL

images_data = soup.find_all('img', class_='s-image')
#print(images_data)

images_url = []
for item in images_data:
    images_url.append(item['src'])
    
images_url=images_url[5:]
images_url, len(images_url)
```

```python
# getting the average rating of the products

rating_data = soup.find_all('span', class_='a-icon-alt')[4:20]
# print(rating_data, len(rating_data))

ratings_list = []

for rating in rating_data:
    ratings_list.append(rating.text[0:3])
    
    
ratings_list, len(ratings_list)
```

```python
amazon_mobiles_under_20000 = pd.DataFrame()

amazon_mobiles_under_20000['Product'] = products_list 
amazon_mobiles_under_20000['Price'] = prices_list
amazon_mobiles_under_20000['Image URL'] = images_url
amazon_mobiles_under_20000['Rating'] = ratings_list

amazon_mobiles_under_20000
```

<H3>TASK 8 : Write a python program to extract information about the local weather from the National 
Weather Service website of USA, https://www.weather.gov/ for the city, San 
Francisco. You need to extract data about 7 day extended forecast display for the city. 
The data should include period, short description, temperature and description.</H3>

```python
page = requests.get("https://forecast.weather.gov/MapClick.php?lat=37.777120000000025&lon=-122.41963999999996#.YJf1D4fivIU")
soup = BeautifulSoup(page.content)
soup
```

```python
# getting the period data ie day 

period_list = []
period_data = soup.find_all('div', class_='col-sm-2 forecast-label')
#print(period_data)

for count,period in enumerate(period_data):
    if count%2==0:
        period_list.append(period.text)

period_list, len(period_list)

# Here, Today means Sunday 9-May-2021
```

```python
# getting the short description of the weather

short_description_list = []

short_desc = soup.find_all('p', class_='short-desc')
short_desc

for count,short in enumerate(short_desc):
    if count%2==0:
        short_description_list.append(short.text)

short_description_list.append(np.nan)
short_description_list.append(np.nan)
short_description_list, len(short_description_list)
```

```python
# getting the description of the weather

description_list = []
description_data = soup.find_all('div', class_='col-sm-10 forecast-text')
#print(description_data)

for count,desc in enumerate(description_data):
    if count%2==0:
        description_list.append(desc.text)

description_list, len(description_list)
```

```python
# getting the temperature details

temperature_list = []

temperature_data = soup.find_all('p', class_='temp temp-high')
print(temperature_data)

for temp in temperature_data:
    temperature_list.append(temp.text)
        
print(temperature_list, len(temperature_list))

temperature_list.append(np.nan)
temperature_list.append(np.nan)
print(temperature_list)
```

```python
san_francisco_weather = pd.DataFrame()

san_francisco_weather['Day'] = period_list
san_francisco_weather['Description'] = description_list
san_francisco_weather['Short description'] = short_description_list
san_francisco_weather['Temperature'] = temperature_list

san_francisco_weather
```
